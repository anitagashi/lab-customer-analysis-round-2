```python
def clean_data(df):
    df.columns=[e.lower().replace(' ', '_') for e in df.columns]
    df=df.drop(columns=['unnamed:_0','vehicle_type'])
    df=df.dropna()
    df['effective_to_date']=pd.to_datetime(df['effective_to_date'], errors='coerce')
    df['month'] = df['effective_to_date'].dt.month
    return df
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('marketing_customer_analysis.csv') 
display(data)
print(data.shape)
```

```python
# Get numerical and categorical data types

numerical = data.select_dtypes(include=np.number)
categoricals = data.select_dtypes(include=np.object)
```

```python
display(numerical)
```

```python
display(categoricals)
```

```python
# Visualize distributions of numerical variables
for column in numerical.columns:
    sns.displot(data[column])
    plt.show()

```

```python
# Visualize histograms of numerical variables
for column in numerical.columns:
    plt.hist(data[column], bins=10)
    plt.title(column)
    plt.show()
```

```python
# Check multicollinearity between numerical variables
corr_matrix = numerical.corr().abs()
```

```python
# Create a heatmap to visualize correlations
sns.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
# Drop highly correlated features (correlation greater than 0.9)

upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(np.bool))
to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.9)]
numerical = numerical.drop(to_drop, axis=1)
```
